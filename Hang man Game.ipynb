```python
import random
import time

animal = ['cat','dog','cattle','donkey','goat','horse','rabbit']

fruit = ['banana','apple','watermelon','cherry','grape','kiwi']

drink = ['coffee','water','soda','juice','cider','tea']

food = ['salad','spaghetti','rice','steak','cookie']

color = ['red','orange','yellow','green','blue','purple']

while True:
    chances = 6
    user = input("Do you want to play hang man?(y/n): ")
    if user == 'y' or user == 'yes' or user == 'Yes' or user == 'Y':
        genre = random.choice(['animal','fruit','drink','food','color'])
#         print(genre)
        print("The word is related to -> \033[94m%s\033[0m <-" % (genre))
        
        if genre == "animal":
            genre = animal
        elif genre == 'fruit':
            genre = fruit
        elif genre == 'drink':
            genre = drink
        elif genre == 'food':
            genre = food
        elif genre == 'color':
            genre = color
            
        word = random.choice(genre)
#         print(word)
        
        print("\033[94mThe game starts now!\033[0m")
        
        unknown = []
        for x in range(len(word)):
            unknown.append("__ ")
        
        
        while True:
            if chances == 6:
                print("____")
                print("|   |")
                print("|    ")
                print("|    ")
                print("|    ")
                print("|_______")
            elif chances == 5:
                print("____")
                print("|   |")
                print("|   o")
                print("|    ")
                print("|    ")
                print("|_______")
            elif chances == 4:
                print("____")
                print("|   |")
                print("|   o")
                print("|   |")
                print("|    ")
                print("|_______")
            elif chances == 3:
                print("____")
                print("|   |")
                print("|   o")
                print("|  /|")
                print("|    ")
                print("|_______")
            elif chances == 2:
                print("____")
                print("|   |")
                print("|   o")
                print('|  /|\\')
                print("|    ")
                print("|_______")
            elif chances == 1:
                print("____")
                print("|   |")
                print("|   o")
                print("|  /|\\")
                print("|  / ")
                print("|_______")
                
            print("->"," ".join(unknown),"<-")
            
            guess = input("Enter a letter or guess the word: ")
            
            letters = list(word)
            if len(guess) > 1:
                if guess == word:
                    print("You win!")
                    break
                else:
                    print("***************************************************")
                    print("Not the right answer!")
                    chances -= 1
                    
            elif len(guess) == 1:
                if guess in letters:
                    for num in range(len(letters)):
                        if guess == letters[num]:
                            unknown[num] = guess
                    print("***************************************************")
                    print("Nice!")
                else:
                    print("***************************************************")
                    print("Not the right answer!")
                    chances -= 1
                    
            if chances == 0:
                print("____")
                print("|   |")
                print("|   o")
                print("|  /|\\")
                print("|  / \\")
                print("|_______")
                print("You lose! Game Over.")
                break
                
                

    elif user == 'n' or user == 'no' or user == 'No'or user == 'No':
        print("Goodbye!")
        break
    else:
        print("invalid input, try again!")
```
